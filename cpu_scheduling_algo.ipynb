```python
print("$ python3 bankers.py")

# Input
n = int(input("Enter number of processes: "))
m = int(input("Enter number of resources: "))

# Allocation matrix
alloc = []
print("Enter Allocation Matrix:")
for i in range(n):
    while True:
        row = list(map(int, input(f"Row {i}: ").split()))
        if len(row) == m:
            alloc.append(row)
            break
        else:
            print(f"❌ Enter exactly {m} values")

# Max matrix
maxm = []
print("Enter Max Matrix:")
for i in range(n):
    while True:
        row = list(map(int, input(f"Row {i}: ").split()))
        if len(row) == m:
            # check allocation <= max
            valid = True
            for j in range(m):
                if row[j] < alloc[i][j]:
                    print(f"❌ Allocation > Max at P{i}, Resource {j}")
                    valid = False
                    break
            if valid:
                maxm.append(row)
                break
        else:
            print(f"❌ Enter exactly {m} values")

# Available resources
while True:
    avail = list(map(int, input("Enter Available Resources: ").split()))
    if len(avail) == m:
        break
    else:
        print(f"❌ Enter exactly {m} values")

# Need matrix
need = []
for i in range(n):
    temp = []
    for j in range(m):
        temp.append(maxm[i][j] - alloc[i][j])
    need.append(temp)

print("\nNeed Matrix:")
for i in range(n):
    print(need[i])

# Safety Algorithm
finish = [False] * n
safeSeq = []
work = avail.copy()

while len(safeSeq) < n:
    found = False
    for i in range(n):
        if not finish[i]:
            if all(need[i][j] <= work[j] for j in range(m)):
                for j in range(m):
                    work[j] += alloc[i][j]
                safeSeq.append(i)
                finish[i] = True
                found = True
    if not found:
        break

# Output
if len(safeSeq) == n:
    print("\nSystem is in SAFE state")
    print("Safe sequence:", end=" ")
    for i in safeSeq:
        print(f"P{i}", end=" ")
else:
    print("\nSystem is NOT in safe state")
```
